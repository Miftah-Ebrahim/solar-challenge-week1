```python
import pandas as pd
import matplotlib.pyplot as plt

# âœ… Load Togo CSV (tab-delimited)
df = pd.read_csv("../data/togo.csv", sep="\t")  # use sep="\t"

# Preview columns and first 5 rows
print(df.columns.tolist())
print(df.head())

# Clean column names (remove extra whitespace if any)
df.columns = [c.strip() for c in df.columns]
print(df.columns.tolist())
```

```python
# Check dataframe info
print(df.info())

# Check for missing values
print(df.isna().sum())
```

```python
# Ensure numeric columns are numeric
df["YEAR"] = df["YEAR"].astype(int)
df["MO"] = df["MO"].astype(int)
df["DY"] = df["DY"].astype(int)
df["T2M"] = df["T2M"].astype(float)
df["WS10M_MIN"] = df["WS10M_MIN"].astype(float)
```

```python
# Summary stats
print(df.describe())

# Any missing values left
print(df.isna().sum())
```

```python
# Line plot for temperature
plt.figure(figsize=(12, 5))
plt.plot(df.index, df["T2M"], label="Temperature")
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")
plt.title("Daily Temperature in Togo")
plt.legend()
plt.show()

# Line plot for wind speed
plt.figure(figsize=(12, 5))
plt.plot(df.index, df["WS10M_MIN"], color="orange", label="Wind Speed")
plt.xlabel("Date")
plt.ylabel("Wind Speed (m/s)")
plt.title("Daily Minimum Wind Speed in Togo")
plt.legend()
plt.show()
```

```python
# Save cleaned datasets:
df.to_csv("../data/togo_clean.csv", index=False)
```
